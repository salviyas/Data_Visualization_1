```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.plot(x,y)
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
categories=["A","B","C","D"]
values=[5,7,3,8]

plt.bar(categories,values)
plt.title("Bar plot")
plt.xlabel("categories")
plt.ylabel("valles")
```

```python
# Box plots are used to visualize the distribution of numerical data,
# particularly showing the quartiles, median, and potential outliers.

# Key components of a box plot:

# - The box itself represents the interquartile range (IQR), which contains the middle 50% of the data.
# - The line inside the box represents the median (50th percentile).
# - The "whiskers" extend from the box to show the range of the data, excluding outliers.
# - Outliers are plotted as individual points beyond the whiskers.

# Example using the 'tips' dataset:
sns.boxplot(x="day",y="total_bill",data=tips)
```
